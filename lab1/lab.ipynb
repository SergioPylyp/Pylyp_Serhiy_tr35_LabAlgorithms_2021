```python
def eratosthenes(N):    
    a = [i for i in range(1, N+1)] 
    b = []
    num=2 
    i=0 
    a.remove(1) 
    
    while num < a[-1]: 
        while len(a) > i:
            if a[i]%num == 0 and a[i]!= num: 
                b.append(a[i]) 
                a.pop(i) 
            i+=1 
        i=0 
 
        
        for j in range(len(a)):
            if a[j] > num:
                num = a[j]
                break
 
    return b

```

```python
import time
import memory_profiler

yAxisMemory = []
yAxisTime = []
xAxis = []

i=0
k=10000
while i < 50:
    xAxis.append(i)
    m1 = memory_profiler.memory_usage()
    
    start_time = time.time()
    eratosthenes(k)
    yAxisTime.append(time.time() - start_time)
    m2=memory_profiler.memory_usage() 
    yAxisMemory.append(m2[0] - m1[0])
    print(f"It took {yAxisTime[i]} Secs and {yAxisMemory[i]} Mb ")

    i=i+1
    
input

%matplotlib notebook
import matplotlib.pyplot as plt
plt.subplot(2, 2, 1)
plt.plot(xAxis, yAxisTime)

plt.subplot(2, 2, 3)
plt.plot(xAxis, yAxisMemory)
```
