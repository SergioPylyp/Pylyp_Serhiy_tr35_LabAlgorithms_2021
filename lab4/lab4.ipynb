```python
import numpy as np
from time import process_time 
import os, psutil, math
import matplotlib.pyplot as plt
import copy
from numpy import random
import numpy as np

t1 = 0
x = random.randint(1000000, size=(50))

#print(x)
process = psutil.Process(os.getpid())
def shellSort(data):
    length=len(data)
    gap = length//2
    while gap > 0:            
        for iIndex in range(gap, length):
                temp = data[iIndex]
                jIndex = iIndex
                while jIndex >= gap and data[jIndex - gap] > temp:                    
                    data[jIndex] = data[jIndex - gap]
                    jIndex = jIndex - gap
                data[jIndex] = temp
        gap =gap//2
    #print(data)
    return data
    
def main():
    print('ORIGINAL LIST:')
    #print(input_list)
    best=100000
    worst=0
    Total=0
    for i in range(1000):
        input_list = x
        t1_start = process_time()
        sorted_list = shellSort(input_list)
        t1_stop = process_time()
        t=t1_stop-t1_start
        if (t<best): best=t
        if (t>worst): worst=t
        Total+=t
    
    print("Best time - ",best)
    print("Worst time - ",worst)
    print("Avarage time - ",Total/1000)
    
main()


fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot([50,100,500,1000,5000,10000,50000,100000], [0.015625,0.015625,0.03125,0.015625,0.046875,0.109375,0.796875,1.796875], label='Worst-case')  # Plot some data on the (implicit) axes.
ax.plot([50,100,500,1000,5000,10000,50000,100000], [0.0,0.0001875,0.001328125,0.00296875,0.02065625,0.044890625,0.26246875,0.528390625], label='Average-case')  # etc.
ax.plot([50,100,500,1000,5000,10000,50000,100000], [0,0,0,0.0,0.0,0.015625,0.234375,0.5], label='Best-case')
ax.set_xlabel('Elements')
ax.set_ylabel('Time, sec')
ax.legend()



#Memory comparison
k = np.linspace(0, 100000, 100)
fig, ak = plt.subplots()  # Create a figure and an axes.
#ak.plot(k, k, label='Experimental memory usage')  # Plot some data on the (implicit) axes.
ak.plot([50,100,500,1000,5000,10000,50000,100000], [110.403584,107.675648,96.70656,104.525824,98.455552,101.732352,99.65568,96.538624], label='Practical memory usage')  # etc.
ak.set_xlabel('Elements')
ak.set_ylabel('Mmeory, Mbits')
ak.set_title("Memory comparison")
ak.legend()

#-------------------------------------
print("Memory Usage:", (process.memory_info().rss/1000000))  # in Mbits
    
```
